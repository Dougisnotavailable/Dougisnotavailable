```python
import pandas as pd        # Para manipular e gerar Dataframes , onde toda maracutaia se estabelece
from google.colab import files    # Para importar os arquivos que estão no meu disco (local)
import io                         # Para importar os arquivos que estão no meu disco (local)
from dateutil import parser   # Para tentar corrigir as datas
```

```python
# Ambos os arquivos contém muitos acentos e estão em Português, se der alguma zebra para ler usar preferencialmente encoding ”utf-8”

DF_RelatorioPJ = files.upload()
df_pj = pd.read_csv(io.BytesIO(DF_RelatorioPJ['Relatório PJ Sirius.xlsx - Relatório_20_04_2023_15_53_50.csv']))
```

```python
DF_ClientesPF = files.upload()
df_pf = pd.read_csv(io.BytesIO(DF_ClientesPF['Clientes PF - CDL.xlsx - Clientes PF.csv']))
```

```python
df_pj.head()
```

```python
# Corrigindo as Datas

df_pj['DataInicio'] = df_pj['DataInicio'].apply(lambda x: parser.parse(str(x)))
df_pj['DataTermino'] = df_pj['DataTermino'].apply(lambda x: parser.parse(str(x)))

df_pf['Data Inicio do contrato'] = df_pf['Data Inicio do contrato'].apply(lambda x: parser.parse(str(x)))
df_pf['Data Término do contrato'] = df_pf['Data Término do contrato'].apply(lambda x: parser.parse(str(x)))

df_pf.head(10)
```

```python
# Corrigindo typos

df_pj['Marca'] = df_pj['Marca'].replace('Vokswagem', 'Volkswagen')
df_pj['Marca'] = df_pj['Marca'].replace('Renaut', 'Renault')

df_pf = df_pf.rename(columns={'Unnamed: 7': 'Marca'})
df_pf['Marca'] = df_pf['Marca'].replace('Vokswagem', 'Volkswagen')
df_pf['Marca'] = df_pf['Marca'].replace('FiatFi', 'Fiat')
```

```python
df_pj.to_csv('pj.csv', index=False)
df_pf.to_csv('pf.csv', index=False)
```
